```python
#12 Create a set of about 3 elements

hw_set = {'Banana', 'Orange', 'Apple', 'Pear', 'Grape'}
print(hw_set)
hw_set.add('Toyota')
print(hw_set)

hw_set2 = {'Tank', 'Ship', 'Airplane'}

hw_unionset = hw_set.union(hw_set2)
print(hw_unionset)

#Pop an element from 2nd set_
hw_set2.pop()
print(hw_set2)

#Clear 1st set
hw_set.clear()
print("has the set been cleared...",hw_set)

#Discard an element and Remove another then print remaining set.
hw_unionset.discard("Ship")
hw_unionset.remove("Tank")
print(hw_unionset)
```

```python
#13 Create a dictionary with at least 5 values of different data types

nba_player = {
    "player_name" : "Lebron James",
    "team" : "Lakers",
    "age" : 38,
    "jersey" : 6,
    "city_born" : "Akron",
    "state_born" : "Ohio"
}

#Retrieve one item from the Dictionary
player_team = nba_player.get("team")
print(player_team)

#Replace one element with your name
nba_player.update({"player_name" : "Sammy Casillas"})
print(nba_player)

#Add my favorite color to dictionary
nba_player.update({"favorite_color": "Blue"})
print(nba_player)

#Add tuple, list, or set to your dictionary
nba_player.update({"player_name":["Patrick Ewing", "Dennis Rodman", "Shaq O'Neal"]})
print(nba_player)


#Print the dictionary keys and values
k = nba_player.keys()
print(k)

v = nba_player.values()
print(v)


#Copy your dictionary into a 2nd dictionary
nba_player_copy = nba_player.copy()
print(nba_player_copy)

#Pop an element from 2nd dictionary
nba_player_copy.popitem()
print(nba_player_copy)

#Remove all items from 2nd dictionary and print 2nd dictionary
nba_player_copy.clear()
print(nba_player_copy)
```
